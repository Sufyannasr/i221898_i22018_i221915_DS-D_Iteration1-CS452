```python
# Improved visualizations for DEiT training (keeps all training logic identical)
# Saves nicer plots to /kaggle/working/figs/

import os, sys, random, json, subprocess, importlib
import math

def ensure_pkg(name, pip_name=None):
    pip_name = pip_name or name
    try:
        return importlib.import_module(name)
    except Exception:
        print(f"Installing {pip_name} ...")
        subprocess.check_call([sys.executable, "-m", "pip", "install", "--quiet", pip_name])
        return importlib.import_module(name)

# core libs (same as original)
timm = ensure_pkg('timm')
torch = ensure_pkg('torch')
torchvision = ensure_pkg('torchvision')
sklearn = ensure_pkg('sklearn')
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

# sklearn helpers
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.metrics import roc_curve, roc_auc_score, precision_recall_curve, average_precision_score
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

# torch helpers
import torch
import torch.nn as nn
from torchvision import transforms, datasets
from torch.utils.data import DataLoader, Subset

# -------------------------
# (unchanged) Config & reproducibility
# -------------------------
SEED = 42
random.seed(SEED); np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available(): 
    torch.cuda.manual_seed_all(SEED)
    torch.backends.cudnn.benchmark = False
    torch.backends.cudnn.deterministic = True

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", DEVICE)
print("Torch:", torch.__version__, "timm:", timm.__version__)

# Quick-run toggle for debugging
QUICK_RUN = False    # set True to test quickly (fewer epochs / smaller batch)

# Paper hyperparameters (kept as paper: LR multiplicative factor etc.)
BATCH_SIZE = 16 if not QUICK_RUN else 8
LR = 1e-5
WEIGHT_DECAY = 1e-4    # small L2 regularization to reduce overfitting
EPOCHS = 20 if not QUICK_RUN else 3
IMG_SIZE = 224
MUL_LR_FACTOR = 0.995
PATIENCE = 6   # early stopping patience (stop when no val improvement)
VAL_SPLIT = 0.10   # stratified val split from original train (10%)

# paths (user provided)
DATA_ROOT = '/kaggle/input/xraydata/chest_xray'   # must exist, user-provided
SAVE_PATH = '/kaggle/working/best_deit_model_fixed.pth'
FIG_DIR = '/kaggle/working/figs'
os.makedirs(FIG_DIR, exist_ok=True)

# -------------------------
# (unchanged) Validate dataset and counts
# -------------------------
for sub in ['train','val','test']:
    path = os.path.join(DATA_ROOT, sub)
    if sub == 'train' and not os.path.isdir(path):
        raise FileNotFoundError(f"Expected train folder not found at: {path}")
    if sub == 'test' and not os.path.isdir(path):
        raise FileNotFoundError(f"Expected test folder not found at: {path}")

print("Dataset root exists:", DATA_ROOT)
print("Summary of provided folders (counts):")
for split in ['train','val','test']:
    p = os.path.join(DATA_ROOT, split)
    if os.path.isdir(p):
        counts = {}
        for c in sorted(os.listdir(p)):
            cpath = os.path.join(p,c)
            if os.path.isdir(cpath):
                counts[c] = len([f for f in os.listdir(cpath) if os.path.isfile(os.path.join(cpath,f))])
        print(f"  {split}: {counts}")
    else:
        print(f"  {split}: NOT PRESENT")

# -------------------------
# (unchanged) Transforms and ImageFolder (for train we will create a Subset for split)
# -------------------------
train_transform = transforms.Compose([
    transforms.RandomResizedCrop(IMG_SIZE, scale=(0.8,1.0)),  # a bit stronger than simple resize to reduce overfit
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomRotation(10),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225]),
])
eval_transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225]),
])

# Use ImageFolder for the original train; we'll split indices below
orig_train_folder = os.path.join(DATA_ROOT, 'train')
orig_val_folder   = os.path.join(DATA_ROOT, 'val')   # this is tiny in your uploaded dataset, we'll ignore for validation
test_folder       = os.path.join(DATA_ROOT, 'test')

full_train_ds_for_split = datasets.ImageFolder(orig_train_folder, transform=eval_transform)  # use eval transform for consistent indexing
print("Full train dataset size (for splitting):", len(full_train_ds_for_split))
print("Classes mapping:", full_train_ds_for_split.class_to_idx)

# Build lists of indices and labels for stratified split
labels = np.array([y for _, y in full_train_ds_for_split.samples])
indices = np.arange(len(labels))
sss = StratifiedShuffleSplit(n_splits=1, test_size=VAL_SPLIT, random_state=SEED)
train_idx, val_idx = next(sss.split(indices, labels))
print(f"Stratified split: {len(train_idx)} train, {len(val_idx)} val (from original train)")

# Create Subset datasets with the proper transforms
train_ds = Subset(datasets.ImageFolder(orig_train_folder, transform=train_transform), train_idx)
val_ds   = Subset(datasets.ImageFolder(orig_train_folder, transform=eval_transform), val_idx)
test_ds  = datasets.ImageFolder(test_folder, transform=eval_transform)  # keep original test as-is

# Create dataloaders
train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True, num_workers=4, pin_memory=True)
val_loader   = DataLoader(val_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=4, pin_memory=True)
test_loader  = DataLoader(test_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=4, pin_memory=True)

# Print sizes & class balance
def ds_class_counts(ds, name):
    ys = []
    if isinstance(ds, Subset):
        for i in ds.indices:
            ys.append(full_train_ds_for_split.samples[i][1])
    else:
        for _, y in ds.samples:
            ys.append(y)
    unique, counts = np.unique(ys, return_counts=True)
    mapping = dict(zip(unique, counts))
    print(f"{name} class counts:", mapping)


ds_class_counts(train_ds, "Train")
ds_class_counts(val_ds, "Val")
ys_test = [y for _, y in test_ds.samples]
print("Test class counts:", dict(zip(*np.unique(ys_test, return_counts=True))))

# Identify positive class index (PNEUMONIA)
if 'PNEUMONIA' in full_train_ds_for_split.class_to_idx:
    pos_class_idx = full_train_ds_for_split.class_to_idx['PNEUMONIA']
else:
    pos_class_idx = 1
print("Positive class index (PNEUMONIA):", pos_class_idx)

# -------------------------
# (unchanged) Model (DEiT base) + replace head + L2 reg
# -------------------------
model = timm.create_model('deit_base_patch16_224', pretrained=True)

# Replace classifier head with a 2-class head with dropout to reduce overfitting
import torch.nn as nn
if hasattr(model, 'head') and isinstance(model.head, nn.Linear):
    in_feats = model.head.in_features
    model.head = nn.Sequential(nn.Dropout(0.3), nn.Linear(in_feats, 2))
elif hasattr(model, 'fc') and isinstance(model.fc, nn.Linear):
    in_feats = model.fc.in_features
    model.fc = nn.Sequential(nn.Dropout(0.3), nn.Linear(in_feats, 2))
else:
    model.reset_classifier(num_classes=2, global_pool='avg')
model = model.to(DEVICE)

# -------------------------
# (unchanged) Weighted loss for class imbalance
# -------------------------
# compute class weights from train indices
train_labels = np.array([full_train_ds_for_split.samples[i][1] for i in train_idx])
unique, counts = np.unique(train_labels, return_counts=True)
class_counts = dict(zip(unique, counts))
print("Train labels counts for weighting:", class_counts)
# ensure ordering matches class indices (0,1)
count0 = class_counts.get(0, 0)
count1 = class_counts.get(1, 0)
# weight = inverse frequency
w0 = 1.0 / (count0 + 1e-12)
w1 = 1.0 / (count1 + 1e-12)
weights = torch.tensor([w0, w1], dtype=torch.float).to(DEVICE)
# normalize weights to sum to 2 (optional)
weights = weights / weights.sum() * 2.0
print("Class weights (used in CrossEntropyLoss):", weights.cpu().numpy())

criterion = nn.CrossEntropyLoss(weight=weights)
optimizer = torch.optim.Adam(model.parameters(), lr=LR, weight_decay=WEIGHT_DECAY)
scheduler = torch.optim.lr_scheduler.MultiplicativeLR(optimizer, lr_lambda=lambda epoch: MUL_LR_FACTOR)

# -------------------------
# (unchanged) Evaluation helpers (fixed AUC) and plotting (EXTENDED)
# -------------------------
def collect_scores_and_labels(model, loader, device, pos_class_idx=1):
    model.eval()
    scores_list, probs_list, preds_argmax_list, labels_list = [], [], [], []
    with torch.no_grad():
        for imgs, labels in loader:
            imgs = imgs.to(device)
            outputs = model(imgs)                       # logits
            logits = outputs.cpu().numpy()
            scores = logits[:, pos_class_idx]           # raw logit for pos class
            exps = np.exp(logits - np.max(logits, axis=1, keepdims=True))
            probs = exps[:, pos_class_idx] / np.sum(exps, axis=1)
            preds_argmax = np.argmax(logits, axis=1)
            scores_list.append(scores); probs_list.append(probs)
            preds_argmax_list.append(preds_argmax); labels_list.append(labels.numpy())
    scores = np.concatenate(scores_list)
    probs = np.concatenate(probs_list)
    preds_argmax = np.concatenate(preds_argmax_list)
    labels = np.concatenate(labels_list)
    preds_prob05 = (probs >= 0.5).astype(int)
    return scores, probs, preds_prob05, preds_argmax, labels


def compute_metrics(labels, preds, scores_or_probs):
    acc = accuracy_score(labels, preds)
    prec = precision_score(labels, preds, zero_division=0)
    rec = recall_score(labels, preds, zero_division=0)
    f1 = f1_score(labels, preds, zero_division=0)
    try:
        auc = float(roc_auc_score(labels, scores_or_probs)) if len(np.unique(labels))>1 else float('nan')
    except:
        auc = float('nan')
    try:
        ap = float(average_precision_score(labels, scores_or_probs)) if len(np.unique(labels))>1 else float('nan')
    except:
        ap = float('nan')
    tn, fp, fn, tp = confusion_matrix(labels, preds).ravel()
    specificity = tn / (tn + fp) if (tn + fp) > 0 else 0.0
    return {'acc':acc, 'prec':prec, 'recall':rec, 'f1':f1, 'auc':auc, 'ap':ap, 'specificity':specificity, 'confusion':(int(tn),int(fp),int(fn),int(tp))}

# New: richer plotting utilities

def plot_roc_pr_combined(labels, scores, probs, title_prefix="", savepath=None):
    if len(np.unique(labels))<2:
        print(f"{title_prefix}: only one class present; cannot plot ROC/PR.")
        return
    fpr, tpr, thr_roc = roc_curve(labels, scores)
    auc = roc_auc_score(labels, scores)
    prec, rec, thr_pr = precision_recall_curve(labels, probs)
    ap = average_precision_score(labels, probs)

    # compute Youden's J for optimal ROC threshold
    j_scores = tpr - fpr
    j_idx = np.argmax(j_scores)
    opt_threshold = thr_roc[j_idx] if j_idx < len(thr_roc) else None

    fig, axes = plt.subplots(1,2, figsize=(12,5))
    axes[0].plot(fpr, tpr, label=f'ROC (AUC={auc:.4f})')
    axes[0].plot([0,1],[0,1],'--')
    if opt_threshold is not None:
        axes[0].scatter(fpr[j_idx], tpr[j_idx], marker='o')
        axes[0].annotate(f"opt thr={opt_threshold:.3f}", (fpr[j_idx], tpr[j_idx]))
    axes[0].set_xlabel('False Positive Rate'); axes[0].set_ylabel('True Positive Rate'); axes[0].set_title(f'{title_prefix} ROC'); axes[0].legend(); axes[0].grid(True)

    axes[1].plot(rec, prec, label=f'PR (AP={ap:.4f})')
    axes[1].set_xlabel('Recall'); axes[1].set_ylabel('Precision'); axes[1].set_title(f'{title_prefix} Precision-Recall'); axes[1].legend(); axes[1].grid(True)

    plt.tight_layout()
    if savepath:
        fig.savefig(savepath, dpi=150)
    plt.show()


def plot_confusion_matrix_heatmap(labels, preds, class_names, title_prefix="", savepath=None):
    cm = confusion_matrix(labels, preds)
    tn, fp, fn, tp = cm.ravel()
    cm_percent = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    fig, ax = plt.subplots(figsize=(5,4))
    im = ax.imshow(cm, interpolation='nearest', cmap='viridis')
    ax.set_title(f"{title_prefix} Confusion Matrix")
    tick_marks = np.arange(len(class_names))
    ax.set_xticks(tick_marks); ax.set_yticks(tick_marks)
    ax.set_xticklabels(class_names); ax.set_yticklabels(class_names)
    ax.set_ylabel('True label'); ax.set_xlabel('Predicted label')

    # annotate counts and percentages
    thresh = cm.max() / 2.
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            count = cm[i,j]
            pct = cm_percent[i,j]
            ax.text(j, i, f"{count}\n({pct:.2%})",
                    ha="center", va="center",
                    color="white" if cm[i, j] > thresh else "black")
    fig.colorbar(im, ax=ax)
    plt.tight_layout()
    if savepath:
        fig.savefig(savepath, dpi=150)
    plt.show()


def plot_training_curves(train_losses, train_accs, val_losses, val_aucs, savepath=None):
    epochs = np.arange(1, len(train_losses)+1)
    fig, ax1 = plt.subplots(figsize=(8,4))
    ax1.plot(epochs, train_losses, label='Train loss')
    if val_losses:
        ax1.plot(epochs, val_losses, label='Val loss')
    ax1.set_xlabel('Epoch'); ax1.set_ylabel('Loss')
    ax1.grid(True)

    ax2 = ax1.twinx()
    if train_accs:
        ax2.plot(epochs, train_accs, label='Train acc', linestyle='--')
    if val_aucs:
        ax2.plot(epochs, val_aucs, label='Val AUC', linestyle=':')
    ax2.set_ylabel('Acc / AUC')

    # combine legends
    lines, labels = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax1.legend(lines+lines2, labels+labels2, loc='best')
    plt.title('Training curves')
    fig.tight_layout()
    if savepath:
        fig.savefig(savepath, dpi=150)
    plt.show()

# helper to unnormalize and show images (Tensor)
def imshow_tensor(img_tensor, ax=None, title=None):
    # img_tensor is CHW normalized
    mean = np.array([0.485,0.456,0.406])
    std = np.array([0.229,0.224,0.225])
    img = img_tensor.numpy().transpose((1,2,0))
    img = (img * std) + mean
    img = np.clip(img, 0, 1)
    if ax is None:
        plt.imshow(img); plt.axis('off')
        if title: plt.title(title)
    else:
        ax.imshow(img); ax.axis('off')
        if title: ax.set_title(title)


def show_sample_predictions(model, dataset, device, class_names, n=12, savepath=None):
    # draw n random samples from dataset and show true/pred/prob
    model.eval()
    # build indices for dataset (handles Subset)
    if isinstance(dataset, Subset):
        idxs = random.sample(list(dataset.indices), min(n, len(dataset.indices)))
    else:
        idxs = random.sample(range(len(dataset)), min(n, len(dataset)))

    fig_cols = 4
    fig_rows = math.ceil(len(idxs) / fig_cols)
    fig, axes = plt.subplots(fig_rows, fig_cols, figsize=(fig_cols*3, fig_rows*3))
    axes = np.array(axes).reshape(-1)

    for ax_i, idx in enumerate(idxs):
        if isinstance(dataset, Subset):
            img_path, label = dataset.dataset.samples[idx]
            img, _ = dataset.dataset[idx]
        else:
            img_path, label = dataset.samples[idx]
            img, _ = dataset[idx]

        with torch.no_grad():
            inp = img.unsqueeze(0).to(device)
            out = model(inp)
            logits = out.cpu().numpy()
            exps = np.exp(logits - np.max(logits, axis=1, keepdims=True))
            probs = exps[:, pos_class_idx] / np.sum(exps, axis=1)
            pred = int(np.argmax(logits, axis=1)[0])
            prob = float(probs[0])

        title = f"T:{class_names[label]}\nP:{class_names[pred]} ({prob:.2f})"
        imshow_tensor(img.cpu(), ax=axes[ax_i], title=title)
        # border color to indicate correct/incorrect
        if pred == label:
            axes[ax_i].spines['left'].set_edgecolor('green'); axes[ax_i].spines['left'].set_linewidth(3)
        else:
            axes[ax_i].spines['left'].set_edgecolor('red'); axes[ax_i].spines['left'].set_linewidth(3)

    # hide leftover axes
    for j in range(ax_i+1, len(axes)):
        axes[j].axis('off')

    plt.tight_layout()
    if savepath:
        fig.savefig(savepath, dpi=150)
    plt.show()

# -------------------------
# (unchanged) Train + early stopping (monitor val AUC)
# -------------------------
best_val_auc = -1.0
best_epoch = -1
no_improve = 0

# new trackers for visualization
train_losses = []
train_accs  = []
val_losses   = []
val_aucs     = []

print("Starting training: epochs =", EPOCHS, "batch =", BATCH_SIZE)
for epoch in range(1, EPOCHS+1):
    model.train()
    running_loss = 0.0
    all_preds = []
    all_targets = []
    for imgs, targets in tqdm(train_loader, desc=f"Epoch {epoch} train", leave=False):
        imgs = imgs.to(DEVICE); targets = targets.to(DEVICE)
        optimizer.zero_grad()
        outputs = model(imgs)
        loss = criterion(outputs, targets)
        loss.backward()
        # gradient clipping to stabilize
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=2.0)
        optimizer.step()
        running_loss += loss.item() * imgs.size(0)
        preds = outputs.argmax(dim=1).detach().cpu().numpy()
        all_preds.append(preds); all_targets.append(targets.detach().cpu().numpy())

    train_loss = running_loss / len(train_loader.dataset)
    train_preds = np.concatenate(all_preds); train_targets = np.concatenate(all_targets)
    train_acc = accuracy_score(train_targets, train_preds)

    # evaluate on val set (collect scores)
    scores_val, probs_val, preds_prob05_val, preds_argmax_val, labels_val = collect_scores_and_labels(model, val_loader, DEVICE, pos_class_idx=pos_class_idx)
    # compute metrics (we use probs/scores for AUC)
    val_metrics_argmax = compute_metrics(labels_val, preds_argmax_val, scores_val)
    val_metrics_prob05  = compute_metrics(labels_val, preds_prob05_val, probs_val)

    # compute val loss explicitly for plotting
    model.eval()
    val_running_loss = 0.0
    with torch.no_grad():
        for imgs_v, targets_v in val_loader:
            imgs_v = imgs_v.to(DEVICE); targets_v = targets_v.to(DEVICE)
            outputs_v = model(imgs_v)
            loss_v = criterion(outputs_v, targets_v)
            val_running_loss += loss_v.item() * imgs_v.size(0)
    val_loss = val_running_loss / len(val_loader.dataset) if len(val_loader.dataset)>0 else float('nan')

    # append trackers
    train_losses.append(train_loss)
    train_accs.append(train_acc)
    val_losses.append(val_loss)
    val_aucs.append(val_metrics_argmax['auc'])

    # check perfect separation (diagnose AUC==1)
    pos_scores = scores_val[labels_val==pos_class_idx] if np.any(labels_val==pos_class_idx) else np.array([])
    neg_scores = scores_val[labels_val!=pos_class_idx] if np.any(labels_val!=pos_class_idx) else np.array([])
    perfect_separation = False
    if pos_scores.size>0 and neg_scores.size>0:
        perfect_separation = (pos_scores.min() > neg_scores.max())

    print(f"\nEpoch {epoch}/{EPOCHS} | Train loss: {train_loss:.4f} | Train acc: {train_acc:.4f}")
    print("Val (argmax) -> acc: {:.4f}, recall: {:.4f}, specificity: {:.4f}, AUC(scores): {:.4f}".format(
        val_metrics_argmax['acc'], val_metrics_argmax['recall'], val_metrics_argmax['specificity'], val_metrics_argmax['auc']
    ))
    print("Val (prob>=0.5) -> acc: {:.4f}, recall: {:.4f}, specificity: {:.4f}, AUC(probs): {:.4f}".format(
        val_metrics_prob05['acc'], val_metrics_prob05['recall'], val_metrics_prob05['specificity'], val_metrics_prob05['auc']
    ))
    print("Val perfect separation (all pos scores > all neg scores)?", perfect_separation)

    # choose monitored metric as val AUC computed from raw scores (more robust than argmax)
    val_auc = val_metrics_argmax['auc']  # already computed on scores
    # step scheduler (paper used multiplicative LR)
    scheduler.step()

    # early stopping and best model save on val_auc improvement
    if val_auc > best_val_auc + 1e-6:
        best_val_auc = val_auc
        best_epoch = epoch
        no_improve = 0
        torch.save({'epoch': epoch, 'model_state_dict': model.state_dict(), 'optimizer_state_dict': optimizer.state_dict()}, SAVE_PATH)
        print(f"Saved best model at epoch {epoch} (val_auc {best_val_auc:.4f}) -> {SAVE_PATH}")
    else:
        no_improve += 1
        print(f"No improvement for {no_improve} epochs (best val_auc {best_val_auc:.4f})")
    if no_improve >= PATIENCE:
        print(f"Early stopping (no improvement in {PATIENCE} epochs).")
        break

print(f"\nTraining finished. Best val AUC: {best_val_auc:.4f} at epoch {best_epoch}")

# -------------------------
# (unchanged) Final evaluation on test set using best model (robust)
# -------------------------
if os.path.exists(SAVE_PATH):
    ckpt = torch.load(SAVE_PATH, map_location=DEVICE)
    model.load_state_dict(ckpt['model_state_dict'])
    print("Loaded best model from:", SAVE_PATH)
else:
    print("No saved best model found; using current model weights.")

# collect scores on test
scores_test, probs_test, preds_prob05_test, preds_argmax_test, labels_test = collect_scores_and_labels(model, test_loader, DEVICE, pos_class_idx=pos_class_idx)
test_metrics_argmax = compute_metrics(labels_test, preds_argmax_test, scores_test)
test_metrics_prob05 = compute_metrics(labels_test, preds_prob05_test, probs_test)

print("\nTest metrics (argmax logits):")
print(json.dumps(test_metrics_argmax, indent=2))
print("\nTest metrics (prob>=0.5):")
print(json.dumps(test_metrics_prob05, indent=2))

# plot ROC/PR for test (improved)
plot_roc_pr_combined(labels_test, scores_test, probs_test, title_prefix="Test", savepath=os.path.join(FIG_DIR,'test_roc_pr.png'))

# plot training curves
plot_training_curves(train_losses, train_accs, val_losses, val_aucs, savepath=os.path.join(FIG_DIR,'training_curves.png'))

# confusion matrix heatmap
class_names = [k for k,v in sorted(full_train_ds_for_split.class_to_idx.items(), key=lambda x:x[1])]
plot_confusion_matrix_heatmap(labels_test, preds_argmax_test, class_names, title_prefix='Test', savepath=os.path.join(FIG_DIR,'test_confusion_matrix.png'))

# class distribution bar charts

def plot_class_distribution(train_ds, val_ds, test_ds, class_map, savepath=None):
    # compute counts
    def counts_from_ds(ds):
        ys = []
        if isinstance(ds, Subset):
            for i in ds.indices:
                ys.append(full_train_ds_for_split.samples[i][1])
        else:
            for _, y in ds.samples:
                ys.append(y)
        unique, counts = np.unique(ys, return_counts=True)
        out = dict(zip(unique, counts))
        return out

    train_counts = counts_from_ds(train_ds)
    val_counts = counts_from_ds(val_ds)
    test_counts = counts_from_ds(test_ds)

    labels = [class_map[k] for k in sorted(class_map, key=lambda x:class_map[x])]
    vals_train = [train_counts.get(i,0) for i in range(len(labels))]
    vals_val = [val_counts.get(i,0) for i in range(len(labels))]
    vals_test = [test_counts.get(i,0) for i in range(len(labels))]

    x = np.arange(len(labels))
    width = 0.23
    fig, ax = plt.subplots(figsize=(6,4))
    ax.bar(x - width, vals_train, width, label='Train')
    ax.bar(x, vals_val, width, label='Val')
    ax.bar(x + width, vals_test, width, label='Test')
    ax.set_xticks(x); ax.set_xticklabels(labels)
    ax.set_ylabel('Count'); ax.set_title('Class distribution (train/val/test)')
    ax.legend()
    plt.tight_layout()
    if savepath:
        fig.savefig(savepath, dpi=150)
    plt.show()

plot_class_distribution(train_ds, val_ds, test_ds, full_train_ds_for_split.class_to_idx, savepath=os.path.join(FIG_DIR,'class_distribution.png'))

# show sample predictions from test
show_sample_predictions(model, test_ds, DEVICE, class_names, n=12, savepath=os.path.join(FIG_DIR,'sample_predictions_test.png'))

# Save test metrics
out_json = '/kaggle/working/test_metrics_fixed_full.json'
with open(out_json, 'w') as f:
    json.dump({'test_argmax': test_metrics_argmax, 'test_prob05': test_metrics_prob05}, f, indent=2)
print("Saved test metrics to", out_json)

print('Saved visualizations to', FIG_DIR)
```
